## Data visualisations using graphs and annotations
Annotations and labelling in a graph leads to better visualisations and we can present the information much more vividly

```python
#import all the libraries needed
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 10,6 # setting the size of the image
```

#### Graphs need to be self-explanatory, labelling and adding legends might further the cause

```python
#lets plot a graph first
a = range(2,11)
b = range(1,10)
plt.bar(a,b)
#setting the labels for the graphs
plt.xlabel('this is a')
plt.ylabel('this is b')
```

```python
#this similar labelling can be extended to pie charts and other charts, lets try :

c = range(1,5)
#1,2,3,4

names = ['first','second','third','fourth']

plt.pie(c, labels = names)

#adding legends for better description
plt.legend(names, loc='best')
plt.show()
```

```python
#note that in this method we are forming an empty graph first and then we are populating it with the data
cars = pd.read_csv('C:/Users/Asus/Desktop/nikhil projects/linkedin/Ex_Files_Python_Data_Science_EssT_Pt_1/Exercise Files/Data/mtcars.csv')
#Note here that the file needs to have the right extension at the end so that path works and the path needs to be in quotes

mpg = cars.mpg

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1]) #[left,bottom,width ,height] range is [0,1]

mpg.plot()

ax.set_xticks(range(32)) #this refers to the different car labels that we have and any other number would generate an error

ax.set_xticklabels(cars.car_names, rotation=80, fontsize='small')
ax.set_title('Miles per Gallon of Cars in mtcars Dataset')

ax.set_xlabel('car names')
ax.set_ylabel('miles/gal')

```
